```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define unwanted substrings or phrases
unwanted_substrings = [
    'client:', 'industry type:', 'organization size:',
    'products & services:', 'summarized:', 'let me know if',
    'according to the og developer', 'developer must',
    'hopefully, with the above notes', 'thank you so much',
    'github account', 'more information from the og developer',
    'do not remove og developer access', 'contact', 'email',
    'whatsapp', 'telegram', 'skype'
]

# Extract article title and cleaned text
def extract_article(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')

        # Get the title
        title_tag = soup.find('title')
        title = title_tag.get_text(strip=True) if title_tag else "No Title Found"

        # Try different ways to get the article content
        article_div = soup.find('div', class_='td-post-content')
        if not article_div:
            article_div = soup.find('article')
        if not article_div:
            article_div = soup.find('div', class_='post-content')

        if not article_div:
            return title, ""

        paragraphs = article_div.find_all('p')
        clean_paragraphs = []
        for p in paragraphs:
            text = p.get_text(strip=True)
            lower_text = text.lower()

            if any(sub in lower_text for sub in unwanted_substrings):
                continue
            if len(text) < 10:
                continue
            if lower_text.startswith('http') or lower_text.startswith('www'):
                continue

            clean_paragraphs.append(text)

        article_text = '\n\n'.join(clean_paragraphs)
        return title, article_text

    except Exception as e:
        print(f"❌ Error extracting from {url}: {e}")
        return None, None

# Save the cleaned article into a text file
def save_article_text(url_id, title, content):
    os.makedirs("extracted_articles", exist_ok=True)
    filename = f"extracted_articles/{url_id}.txt"
    with open(filename, "w", encoding="utf-8") as file:
        file.write(title.strip() + "\n\n")
        file.write(content.strip())

# Load the input DataFrame
input_df = pd.read_excel(r"C:\Users\L.RAMYA\Downloads\Input.xlsx")

# Iterate through URLs and save cleaned articles
for idx, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    title, article_text = extract_article(url)
    if title and article_text:
        save_article_text(url_id, title, article_text)
        print(f"✅ Saved cleaned article for URL_ID: {url_id}")
    else:
        print(f"⚠️ Skipping URL_ID: {url_id} due to extraction failure.")
```

```python
import os
import re
import nltk
import string
import pyphen
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')

# Set paths
article_folder = r"D:\Jupyter\RESUME PROJECTS\extracted_articles"
positive_words_path = r"C:\Users\L.RAMYA\Downloads\MasterDictionary-20250522T160006Z-1-001\MasterDictionary\negative-words.txt"
negative_words_path = r"C:\Users\L.RAMYA\Downloads\MasterDictionary-20250522T160006Z-1-001\MasterDictionary\negative-words.txt"
stopwords_folder = r"C:\Users\L.RAMYA\Downloads\StopWords-20250522T160009Z-1-001"

# Load positive and negative words
def load_words(file_path):
    with open(file_path, 'r', encoding='latin-1') as f:
        return set([line.strip().lower() for line in f if line.strip() and not line.startswith(';')])

positive_words = load_words(positive_words_path)
negative_words = load_words(negative_words_path)

# Load stopwords
stop_words = set()
for filename in os.listdir(stopwords_folder):
    file_path = os.path.join(stopwords_folder, filename)
    if os.path.isfile(file_path):  # ✅ Skip folders
        with open(file_path, 'r', encoding='latin-1') as f:
            stop_words.update([line.strip().lower() for line in f if line.strip()])


# Syllable counter
dic = pyphen.Pyphen(lang='en')

def count_syllables(word):
    hyphenated = dic.inserted(word)
    return max(1, hyphenated.count('-') + 1)

def is_complex(word):
    return count_syllables(word) > 2

def analyze_article(text):
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    words_clean = [w.lower() for w in words if w.isalpha()]
    filtered_words = [w for w in words_clean if w not in stop_words]

    positive_score = sum(1 for w in filtered_words if w in positive_words)
    negative_score = sum(1 for w in filtered_words if w in negative_words)

    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 1e-6)
    subjectivity_score = (positive_score + negative_score) / (len(filtered_words) + 1e-6)

    total_words = len(filtered_words)
    total_sentences = len(sentences)
    avg_sentence_length = total_words / total_sentences if total_sentences != 0 else 0

    complex_words = [w for w in filtered_words if is_complex(w)]
    complex_word_count = len(complex_words)
    percentage_complex_words = complex_word_count / total_words if total_words != 0 else 0

    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

    total_syllables = sum(count_syllables(w) for w in filtered_words)
    syllables_per_word = total_syllables / total_words if total_words != 0 else 0

    # Personal pronouns (basic version)
    personal_pronouns = len(re.findall(r'\b(I|we|my|ours|us)\b', text, flags=re.I))

    avg_word_length = sum(len(w) for w in filtered_words) / total_words if total_words != 0 else 0

    return {
        "POSITIVE SCORE": positive_score,
        "NEGATIVE SCORE": negative_score,
        "POLARITY SCORE": polarity_score,
        "SUBJECTIVITY SCORE": subjectivity_score,
        "AVG SENTENCE LENGTH": avg_sentence_length,
        "PERCENTAGE OF COMPLEX WORDS": percentage_complex_words,
        "FOG INDEX": fog_index,
        "AVG NUMBER OF WORDS PER SENTENCE": avg_sentence_length,
        "COMPLEX WORD COUNT": complex_word_count,
        "WORD COUNT": total_words,
        "SYLLABLE PER WORD": syllables_per_word,
        "PERSONAL PRONOUNS": personal_pronouns,
        "AVG WORD LENGTH": avg_word_length
    }

# Analyze all articles
import pandas as pd

results = []

for filename in os.listdir(article_folder):
    if filename.endswith(".txt"):
        with open(os.path.join(article_folder, filename), 'r', encoding='utf-8') as f:
            lines = f.readlines()
            if len(lines) > 1:
                title = lines[0].strip()
                content = " ".join(lines[1:]).strip()
                scores = analyze_article(content)
                scores["URL_ID"] = filename.replace(".txt", "")
                scores["TITLE"] = title
                results.append(scores)

# Save to Excel with correct column order
df_result = pd.DataFrame(results)

# Correct column order required by the evaluator
correct_column_order = [
    "URL_ID", "TITLE", 
    "POSITIVE SCORE", "NEGATIVE SCORE", "POLARITY SCORE", "SUBJECTIVITY SCORE", 
    "AVG SENTENCE LENGTH", "PERCENTAGE OF COMPLEX WORDS", "FOG INDEX", 
    "AVG NUMBER OF WORDS PER SENTENCE", "COMPLEX WORD COUNT", "WORD COUNT", 
    "SYLLABLE PER WORD", "PERSONAL PRONOUNS", "AVG WORD LENGTH"
]

# Reorder the columns
df_result = df_result[correct_column_order]

# Save the result
df_result.to_excel("Formatted_Output_Data.xlsx", index=False)
print("✅ Final output saved to 'Formatted_Output_Data.xlsx' with correct column format.")
```
